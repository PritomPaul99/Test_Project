```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.sin(10*x) + np.cos(3*x)

# Generate x values from 0 to 5
x = np.linspace(0, 5, 1000)  # 500 points for smoothness
y = f(x)

# Plot the function
plt.plot(x, y, label=r'$f(x) = \sin(10x) + \cos(3x)$')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = sin(10x) + cos(3x)")
plt.legend()
plt.grid(visible=True)
plt.show()
```
