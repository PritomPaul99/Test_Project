```python
import matplotlib.pyplot as plt
import numpy  as np
```

```python
x = np.linspace(1, 100, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x)
y4 = np.exp(x)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y1")
plt.title("sin(x)")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y2")
plt.title("cos(x)")

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.xlabel("x")
plt.ylabel("y3")
plt.title("log(x)")

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.xlabel("x")
plt.ylabel("y4")
plt.title("exp(x)")

# plt.subplots_adjust(wspace=0.5, hspace=0.5) 
# or
plt.tight_layout()  # to remove overlapping plots
plt.suptitle("Different types of functions.")
plt.savefig("Four Plots.png", dpi=500)
plt.show()
```
