```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(c):
    return ((668.06/c)*(1 - np.exp(-0.146843 * c))) - 40
```

```python
x = np.linspace(1, 20, 10000)
y = func(x)
```

```python
res = np.Infinity
c = 0
for  i in x:
    Y = func(i)
    if Y < res and Y >= 0:
        res = Y
        c = i

print(c, int(res))
```

```python
plt.scatter(c, res, color='red', marker='.', s=100, label=f'Min: ({c:.2f}, {res:.2f})')
plt.plot(x, y, label=r"(668.06/c)(1-e^-0.146843c)-40")
plt.plot(0, 0)
plt.xlabel("c")
plt.ylabel("func(c)")
plt.title(r"(668.06/c)(1-e^-0.146843c)-40)")
plt.legend()
plt.grid(True)
plt.show()
```
