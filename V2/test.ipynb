```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 23, 24]

plt.plot(x, y, marker="o", mfc='r', mec='r')
for xi, yi in zip(x, y):
    plt.text(xi, yi, f"({xi}, {yi})", ha='left', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Plot points
plt.plot(x, y, marker='o', linestyle='', color='red')

# Add coordinates above each marker
for xi, yi in zip(x, y):
    plt.text(xi, yi, f'  ({xi}, {yi})',
             fontsize=9, ha='left', va='center')

plt.title("Markers with Coordinates")
plt.show()
```
