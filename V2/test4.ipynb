```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func0(x):
    return np.exp(-x) - x
```

```python
def func1(x):
    return np.exp(-x)
```

```python
def func2(x):
    return x
```

```python
x = np.linspace(0, 1, 100)
y0 = func0(x)
y1 = func1(x)
y2 = func2(x)
```

```python
# plt.figure(1)
from turtle import color
from click import style


plt.figure(figsize=(5, 10), dpi=1000)
plt.subplot(2, 1, 1)
plt.plot(x, y0, label='func0: exp(-x) - x')
plt.plot([0, 1], [0, 0], linestyle='--', color="black")  # x-axis
plt.legend()

# plt.figure(2)
plt.subplot(2, 1, 2)
plt.plot(x, y1, label='func1: exp(-x)')
plt.plot(x, y2, label='func2: x')
plt.plot([0, 1], [0, 0], linestyle="--", color="black")  # x-axis
plt.legend()
plt.show()
```
