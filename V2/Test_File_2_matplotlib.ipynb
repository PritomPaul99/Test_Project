```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(5 * x) + np.cos(2 * x)
```

```python
# pylustrator.start()
plt.figure(1)
plt.plot(x, y1, label="sin(x)", color="green", linestyle="--")
plt.figure(2)
plt.plot(x, y2, label="cos(x)", color="purple", linestyle=":")
plt.plot(x, y3, label="sin(10x) + cos(3x)", color="orange", linestyle="-.")
# plt.plot(np.linspace(0, 10, 1000), np.zeros(1000), color="black")
plt.axis([-5, 15, -3, 3])
plt.title("Different Transcendental functions.")
plt.xlabel("x")
plt.ylabel("y1, y2, y3")
plt.grid()
plt.legend(loc="lower center", fontsize="12", ncol=3)
plt.show()
```
